```python
import matplotlib.pyplot as plt
fig = plt.figure()        
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot([1, 2, 3], [4, 6, 4])
ax.set_title("A Simple Figure")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot([1, 2, 3], [4, 6, 4])
ax.set_title("One plot using plt.subplots()")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 6, 4])
plt.title("One plot using plt.plot()")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(8, 3))

# Left plot
axes[0].plot([1,2,3], [1,4,9])
axes[0].set_title("Quadratic")

# Right plot
axes[1].plot([1,2,3], [1,2,3])
axes[1].set_title("Linear")

plt.tight_layout()  # nice spacing so titles/labels don't overlap
plt.show()
```

```python
import numpy as np
import pandas as pd

x = np.random.rand(5)
y = np.random.rand(5)
#plt.scatter(x, y)
plt.scatter(x, y, c = ['blue', 'green', 'black', 'red', 'maroon'], marker = 's')
plt.title("Scatter Plot")
plt.xlabel("Random X")
plt.ylabel("Random Y")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data for urban area; all data in $k
marketing_urban = np.array([20, 25, 30, 35, 40, 50, 55, 60, 70, 80])
sales_urban = np.array([210,260,300,330,380,480,520,610,700,820])

# data for rural area; all data in $k
marketing_rural = np.array([10, 12, 15, 18, 22, 28, 30, 35, 40, 45])
sales_rural = np.array([90,110,130,150,180,210,230,260,290,330])

fig, ax = plt.subplots(figsize=(7,4))
#Draw both the scatter plot on the same axes 
ax.scatter(marketing_urban, sales_urban, c='red', marker='o', label='Urban')
ax.scatter(marketing_rural, sales_rural, c='orange', marker='^', label='Rural')

ax.set_xlabel('Marketing Expense ($k)')
ax.set_ylabel('Sales ($k)')
ax.set_title('Marketing vs Sales  Urban vs Rural')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D']
values = [23, 45, 12, 35]

plt.bar(categories, values, color='orange')
plt.title("Bar Chart Example")
plt.ylabel("Values")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
