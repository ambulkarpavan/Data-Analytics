```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fig = plt.figure()        
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
#ax = fig.add_axes([0.0, 0.0, 0.5, 0.5]) 
ax.plot([1, 2, 3], [4, 6, 4])
ax.set_title("A Simple Figure")

ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

```python
fig = plt.figure()

# First axes with margin
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax1.set_title("Left=0.1, Bottom=0.1")
ax1.plot([1,2,3], [4,6,4])

# Second axes at corner
ax2 = fig.add_axes([0.0, 0.0, 0.5, 0.5])
ax2.set_title("Left=0.0, Bottom=0.0")
ax2.plot([1,2,3], [1,2,3])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# data for urban area; all data in $k
marketing_urban = np.array([20, 25, 30, 35, 40, 50, 55, 60, 70, 80])
sales_urban = np.array([210,260,300,330,380,480,520,610,700,820])

# data for rural area; all data in $k
marketing_rural = np.array([10, 12, 15, 18, 22, 28, 30, 35, 40, 45])
sales_rural = np.array([90,110,130,150,180,210,230,260,290,330])

fig, ax = plt.subplots(figsize=(7,4))
#Draw both the scatter plot on the same axes 
ax.scatter(marketing_urban, sales_urban, c='red', marker='o', label='Urban')
ax.scatter(marketing_rural, sales_rural, c='orange', marker='^', label='Rural')

ax.set_xlabel('Marketing Expense ($k)')
ax.set_ylabel('Sales ($k)')
ax.set_title('Marketing vs Sales  Urban vs Rural')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 
data = np.random.normal(0, 1, 1000)
# print(data)
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

heights_no_outlier = [160, 162, 163, 165, 166, 167, 168, 168, 169, 170, 171, 172, 173, 174, 175]

heights_with_outlier = [140, 160, 162, 163, 165, 166, 167, 168, 168, 169, 170, 171, 172, 190, 195]

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].boxplot(heights_no_outlier, orientation='vertical')
axes[0].set_title("Class A (No Outliers)")
axes[0].set_ylabel("Height (cm)")

axes[1].boxplot(heights_with_outlier, orientation='vertical')
axes[1].set_title("Class B (With Outliers)")
axes[1].set_ylabel("Height (cm)")

plt.suptitle("Boxplot of Student Heights (2 Classes)", fontsize=14)
plt.show()
```
