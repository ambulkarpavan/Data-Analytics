```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Create a sample DataFrame
data = {
    "Subject": ["Math", "Science", "English", "Math", "Science", "English", "Math", "Science", "English"],
    "Marks": [85, 90, 78, 65, 70, 72, 95, 88, 80],
    "Student_Type": ["Day Scholar", "Hosteller", "Day Scholar", "Hosteller", "Day Scholar", "Hosteller", "Day Scholar", "Hosteller", "Day Scholar"]
}

df = pd.DataFrame(data)
print(df)

# 2. Countplot - distribution of student types
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Student_Type")
plt.title("Student Type Distribution")
plt.show()

# 3. Boxplot - score spread per type
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="Student_Type", y="Marks")
plt.title("Marks Distribution by Student Type")
plt.show()

# 4. Heatmap - correlations of subjects
# First pivot to wide format (subjects as columns)
pivot_df = df.pivot_table(index=df.index, columns="Subject", values="Marks")
corr = pivot_df.corr()

plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Subject Correlation Heatmap")
plt.show()
```
