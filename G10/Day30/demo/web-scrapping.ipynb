```python
# amazon_basic_scraper.py
import requests
from bs4 import BeautifulSoup

# Step 1: Create a headers dictionary (to look like a browser)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/120.0.0.0 Safari/537.36",
    "Accept-Language": "en-US,en;q=0.9"
}

# Step 2: Choose Amazon URL (example: search for mobiles)
url = "https://www.amazon.in/s?k=mobiles&crid=21R2ZL1EMFAJK&sprefix=mobiles%2Caps%2C318&ref=nb_sb_noss_2"

# Step 3: Send HTTP request
response = requests.get(url, headers=headers)

# Step 4: Parse HTML
soup = BeautifulSoup(response.text, 'html.parser')

# Step 5: Find product containers
products = soup.find_all("div", {"data-component-type": "s-search-result"})

# Step 6: Extract name and price
for product in products:
    # product name
    name = product.h2.text.strip() if product.h2 else "N/A"

    # product price
    price_tag = product.find("span", class_="a-price-whole")
    price = price_tag.text.strip() if price_tag else "N/A"

    print(f"Name: {name}")
    print(f"Price: ₹{price}")
    print("-" * 60)
```

```python
import requests
from bs4 import BeautifulSoup
import statistics

# Step 1: URLs to scrape
urls = [
    "http://books.toscrape.com/catalogue/category/books/travel_2/index.html",
    "http://books.toscrape.com/catalogue/category/books/mystery_3/index.html",
    "http://books.toscrape.com/catalogue/category/books/historical-fiction_4/index.html"
]

# Step 2: Rating conversion dictionary
rating_map = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}

# Step 3: Function to scrape products from a single page
def scrape_books(url):
    response = requests.get(url)
    response.encoding = "utf-8"  # Fix encoding issue
    soup = BeautifulSoup(response.text, 'html.parser')
    books = soup.find_all("article", class_="product_pod")

    data = []
    for book in books:
        title = book.h3.a["title"]
        price_text = book.find("p", class_="price_color").text.strip()

        # Clean the price (handles cases like 'Â£45.17' or '£45.17')
        price_text = price_text.replace("Â", "").replace("£", "")
        price_value = float(price_text)

        rating_class = book.find("p", class_="star-rating")["class"][1]
        rating = rating_map[rating_class]

        data.append({
            "title": title,
            "price": price_value,
            "rating": rating,
            "url": url
        })
    return data

# Step 4: Collect all books
all_books = []
for url in urls:
    all_books.extend(scrape_books(url))

# Step 5: Summary statistics
total_products = len(all_books)
prices = [b["price"] for b in all_books]
ratings = [b["rating"] for b in all_books]

min_price = min(prices)
max_price = max(prices)
avg_rating = statistics.mean(ratings)

# Step 6: Lowest price product
lowest_price_product = min(all_books, key=lambda x: x["price"])

# Step 7: Highest rated product
highest_rated_product = max(all_books, key=lambda x: x["rating"])

# Step 8: Value Score Calculation
for book in all_books:
    book["value_score"] = book["rating"] / book["price"]

# Step 9: Top 3 best value products
top_value_books = sorted(all_books, key=lambda x: x["value_score"], reverse=True)[:3]

# Step 10: Print results
print("SUMMARY STATISTICS:")
print(f"Total Products Compared: {total_products}")
print(f"Price Range: £{min_price:.2f} - £{max_price:.2f}")
print(f"Average Rating: {avg_rating:.1f}/5\n")

print("LOWEST PRICE PRODUCT:")
print(f"Title: {lowest_price_product['title']}")
print(f"Price: £{lowest_price_product['price']:.2f}")
print(f"Rating: {lowest_price_product['rating']}/5")
print(f"Source: {lowest_price_product['url']}\n")

print("HIGHEST RATED PRODUCT:")
print(f"Title: {highest_rated_product['title']}")
print(f"Price: £{highest_rated_product['price']:.2f}")
print(f"Rating: {highest_rated_product['rating']}/5")
print(f"Source: {highest_rated_product['url']}\n")

print("TOP 3 BEST VALUE PRODUCTS:")
for i, book in enumerate(top_value_books, start=1):
    print(f"{i}. {book['title']} | Price: £{book['price']:.2f} | Rating: {book['rating']}/5 | URL: {book['url']}")
```

```python
#Do the necessary imports
import requests
from bs4 import BeautifulSoup
import copy
#Create list of URLs
url_list = [
    'http://books.toscrape.com/catalogue/category/books/travel_2/index.html',
    'http://books.toscrape.com/catalogue/category/books/mystery_3/index.html',
    'http://books.toscrape.com/catalogue/category/books/historical-fiction_4/index.html'
]
#create a dictionaty for textual to numerical; format class-name: numerical-value
rating_dict = {'One': 1, 'Two': 2, 'Three': 3, 'Four': 4, 'Five': 5}
#create lists for storing information
title_list = []
price_list = []
rating_list = []
source_list = []
#iterate over each URL
for url in url_list:
    #hit the URL using the get method
    response = requests.get(url)
    #Create an object of BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')
    #get all article tags with class 'product_pod'
    books = soup.find_all('article', class_ = 'product_pod')
    #iterate over every book for current URL
    for book in books:
        #get div that has class 'image_container', access a tag in this div, access img tah in this a, get the value of alt attribute
        title_list.append(book.find('div', class_ = 'image_container').a.img['alt'])
        #get the price
        price_list.append(float(book.find("p", class_ = 'price_color').text[2:]))
        #get the rating
        rating_class = book.find('p', class_ = 'star-rating').get('class')[1]
        rating_list.append(rating_dict[rating_class])
        source_list.append(url)
#print the total number of products to compare
print(f"Total books to compare is {len(title_list)}")
#Price range (minimum and maximum price, formatted to two decimal places)
min_price, max_price = round(min(price_list), 2), round(max(price_list), 2)
print(f"Mininimum Price is £{min_price}, Maximum price is £{max_price}")
#Average rating (mean of all product ratings, formatted to one decimal place)
average_rating = round(sum(rating_list)/len(rating_list) ,1)
print(f"Average rating is {average_rating}")
print()
#The product with the lowest price
#get the index of minimum price
min_index = price_list.index(min(price_list))
print("For book with minimum price")
print(f"Title = {title_list[min_index]}\tPrice = {price_list[min_index]}\tRating = {rating_list[min_index]}\tSource = {source_list[min_index]}")
#get the index of maximum price
max_index = price_list.index(max(price_list))
print("For book with maximum price")
print(f"Title = {title_list[max_index]}\tPrice = {price_list[max_index]}\tRating = {rating_list[max_index]}\tSource = {source_list[max_index]}")
#calculate the value_score using the formula: value_score = rating / price
#create a list to store the value_score
value_score_list = []
#iterate over each element of rating and price
for index, rating in enumerate(rating_list):
    value_score_list.append(round(rating/price_list[index], 2))
print("\n")
#create a deep copy of the value_score_list (name copy_value_score_list)
copy_value_score_list = copy.deepcopy(value_score_list)
#sort the copy_value_score_list in the descnding order
copy_value_score_list = sorted(copy_value_score_list, reverse = True)
#for the first three elements of copy_value_score_list
three_max_value_score = copy_value_score_list[:3]
#get the index of element in the value_score_list that index is useful while fetching the product details
for current_max_value in three_max_value_score:
    #find the index of current_max_value in the value_score_list
    current_index = value_score_list.index(current_max_value)
    print(f"Title = {title_list[current_index]}\tPrice = {price_list[current_index]}\tRating = {rating_list[current_index]}\tSource = {source_list[current_index]}")
```

```python
from bs4 import BeautifulSoup
def classify_format(html: str) -> list:
    soup = BeautifulSoup(html, 'html.parser')
    a_list = soup.find_all(name = "a")
    result = []
    for a in a_list:
        result.append(a.get('href'))
    return result
html = '<html><body><a href="https://example.com">Link</a></body></html>'
result = classify_format(html)
print(result)
print()
html = '<html><body><a href="/home">Home</a><a href="/about">About</a></body></html>'
result = classify_format(html)
print(result)
print()
html = '<html><body><p>No links here</p></body></html>'
result = classify_format(html)
print(result)
```

```python
def classify_format(html: str, tag) -> list:
    soup = BeautifulSoup(html, 'html.parser')
    img_list = soup.find_all(name = tag)
    result = []
    for a in img_list:
        result.append(a.get('src'))
    return result
html = '<html><body><img src="logo.png" /><img src="banner.jpg" /></body></html>'
result = classify_format(html)
print(result)
print()
html = '<html><body><img src="/static/image1.png" /></body></html>'
result = classify_format(html)
print(result)
print()
html = '<html><body><p>No images</p></body></html>'
result = classify_format(html)
print(result)
```
