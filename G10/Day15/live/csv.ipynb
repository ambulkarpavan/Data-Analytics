```python
data = {
    'CustomerID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, np.nan, 35, 40],  # Include a NaN value
    'City': ['Delhi', 'Mumbai', 'Chennai', None],  # None to simulate a missing value
    'Spending': [2500.5, 1800.0, 3200.0, 1500.0]
}

dFrame3 = pd.DataFrame(data)
dFrame3.to_csv('customers.csv', index = False)
print("Check the", os.getcwd(), "for the file customers.csv, especially take care for the missing value" )
```
