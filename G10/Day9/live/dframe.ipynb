```python
ResultSheet={
  'Arnab': pd.Series([90, 91, 97], index=['Maths','Science','Hindi']),
  'Ramit': pd.Series([92, 81, 96], index=['Maths','Science','Hindi']),
  'Samridhi': pd.Series([89, 91, 88], index=['Maths','Science','Hindi']), 
  'Riya': pd.Series([81, 71, 67], index=['Maths','Science','Hindi']),
  'Mallika': pd.Series([94, 95, 99], index=['Maths','Science','Hindi'])
}
dFrame1 = pd.DataFrame(ResultSheet)
print(dFrame1)
#Add column at the end
dFrame1["preeti"] = [89, 90, 91]
print(dFrame1)
dFrame1.insert(1, "Jayesh", np.array([90, 95, 100]))
print(dFrame1)
dFrame1.loc['English'] = [85, 86, 83, 80, 90, 89, 90]
dFrame1.loc['Maths'] = [99, 98, 97, 96, 95, 94, 91]
dFrame1.loc['Hindi']= 0
print(dFrame1)

print()
dFrame1 = dFrame1.rename({'Maths':'Sub1', 'Science':'Sub2','English':'Sub3', 'Hindi':'Sub4'}, axis='index')
dFrame1 = dFrame1.rename({'Arnab':'ST-01', 'Mallika':'ST-02'}, axis='columns')
print(dFrame1)
dFrame1 = dFrame1.drop('Ramit', axis=1)
print(dFrame1)

del dFrame1['ST-01']
print(dFrame1)
```

```python
ar_1 = np.array([10, np.nan, 20, 30, 40])
ar_2 = np.array(['Google', np.nan, 'Meta', 'Amazon', 'Apple'])
ar_3 = np.array([np.nan, {'name': 'Google'}, 123])

print(ar_1.dtype) 
print(ar_2.dtype)
print(ar_3.dtype)

series_1 = pd.Series([1, 2, np.nan], dtype="Int64")
#is same as series_1 = pd.Series([1, 2, np.nan], dtype=pd.Int64Dtype())
print(series_1)

series_2 = pd.Series([1.5, 2.3, np.nan], dtype="Float64")
#is same as series_2 = pd.Series([1.5, 2.3, np.nan], dtype=pd.Float64Dtype())
print(series_2)

series_3 = pd.Series(['New york', 'London', 'Delhi', np.nan], dtype="string")
#is same as series_3 = pd.Series(['New york', 'London', 'Delhi', np.nan], dtype=pd.StringDtype())
print(series_3)

series_5 = pd.Series([{1: "One"}, {2: "Two"}, {3: "Three"}, np.nan], dtype = 'object')
print(series_5)


a = pd.NA
b = pd.NA
print(a == b)
print(a != b)
print(a is b)


series = pd.Series([10, 20, 30, pd.NA], dtype = "Int64")
print(series + 10)
```
