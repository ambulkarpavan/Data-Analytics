```python
import numpy as np
from matplotlib import colormaps

#Get the inbuilt object that for 'viridis' colormap. Viridis is continuous colormap, meaning it smoothly interpolates colors for any value between 0 and 1.
viridis = colormaps['viridis']

# 6 points between 0 and 1
values = np.linspace(0, 1, 6)  
print(values)
#get list of colors from array
colors = [viridis(v) for v in values]
print(colors)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.random.randn(1000)
y = np.random.randn(1000)

# Scatter plot with colormap
sc = plt.scatter(x, y, c=x*y, cmap="viridis")

# Add colorbar (mappable = sc)
plt.colorbar(sc, label="Value (x*y)")

plt.title("Basic Colorbar Example")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure with 2 axes
fig, (ax1, ax2) = plt.subplots(1, 2, gridspec_kw={"width_ratios":[4,1]})

# Create data
data = np.random.rand(10, 10)

# Show image
im = ax1.imshow(data, cmap="plasma")

# Place colorbar inside ax2 (the second axes)
plt.colorbar(im, cax=ax2, label="Intensity")

ax1.set_title("Heatmap with colorbar in ax2")
plt.show()
```

```python
import matplotlib.pyplot as plt

countries = ["Argentina", "Sweden", "South Africa", "Spain", "China", "Uganda", "Japan", "India", 'Mauritius']
densities = [44, 67, 140, 250, 390, 640, 880, 1260, 1650]

# Create scatter plot with viridis colormap
#plt.figure(figsize=(10,6))
plt.subplots(figsize=(10,6))
sc = plt.scatter(countries, densities, c=densities, cmap="viridis", s=200, edgecolor="black", marker="o")
#code to add density as text for marker
for index in range(len(densities)):
  plt.text(countries[index], densities[index] + 50, densities[index], fontsize=8, color = 'black', ha = 'center')

# Add colorbar
cbar = plt.colorbar(sc)
cbar.set_label("Population Density (per km)", fontsize=12)
```

```python
import matplotlib.pyplot as plt
import numpy as np

#generate an array from 0.01, 0.02 ... 10.0
t = np.arange(0.01, 10.0, 0.01)
#find exponent of every number
data1 = np.exp(t)

#create a subplot
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

data2 = np.sin(2 * np.pi * t)
color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
