```python
import matplotlib.pyplot as plt

heights_no_outlier = [160, 162, 163, 165, 166, 167, 168, 168, 169, 170, 171, 172, 173, 174, 175]

heights_with_outlier = [140, 160, 162, 163, 165, 166, 167, 168, 168, 169, 170, 171, 172, 190, 195]
	
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].boxplot(heights_no_outlier, orientation='vertical')
axes[0].set_title("Class A (No Outliers)")
axes[0].set_ylabel("Height (cm)")

axes[1].boxplot(heights_with_outlier, orientation='vertical')
axes[1].set_title("Class B (With Outliers)")
axes[1].set_ylabel("Height (cm)")

plt.suptitle("Boxplot of Student Heights (2 Classes)", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = np.arange(2015, 2025)
admissions = np.array([45, 52, 48, 60, 75, 68, 82, 77, 90, 72])

# Create plot in form of line chart
plt.subplots(figsize=(10, 5))
plt.plot(years, admissions, marker='o', linestyle='-.', linewidth=2, color='tab:blue', label='Admissions')

# Labels, title, legend
plt.xlabel("Year")
plt.ylabel("Number of Admissions")
plt.title("Admissions per Year (20152024)")
plt.legend()

plt.tight_layout()
plt.show()

```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def sinplot(n = 10, flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)

sns.set_theme()   # Apply seaborn defaults
sinplot()
plt.title("Seaborn defaults after sns.set_theme()")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def lineplot(n=3):
    x = np.linspace(0, 10, 100)  # 100 values from 0  10
    
    # Apply seaborn theme (affects background, grid, fonts, etc.)
    sns.set_theme()  
    
    for i in range(1, n + 1):  
        y = x * i   # simple linear function, slope changes with i
        plt.plot(x, y, label=f"Line {i}")  # plot line

    plt.title("Line plots with Seaborn theme")
    plt.xlabel("X-axis values")
    plt.ylabel("Y-axis values")
    plt.legend()
    plt.show()

lineplot()
```
