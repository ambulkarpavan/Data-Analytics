```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style="whitegrid")

# We will use a synthetic dataset:
data = pd.DataFrame({
  'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
  'Math': [88, 92, 79, 93, 85, 70, 89, 76],
  'Science': [85, 90, 78, 88, 86, 72, 87, 80],
  'Grade': ['A', 'A', 'B', 'A', 'B', 'C', 'A', 'B']
})
# barplot - Aggregate categories (mean by default)
sns.barplot(x='Grade', y='Math', data=data)
plt.title("Average Math Score by Grade")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Create the dataframe
tip_data_dict = {
  "total_bill": [16.58, 7.56, 10.34, 43.11, 13.00, 13.51, 18.71, 12.74, 13.00, 16.40, 20.53, 16.47, 26.59, 38.73, 24.27, 12.76, 30.06, 25.89, 48.33, 13.27],
  "tip": [4.00, 1.44, 2.00, 5.00, 2.00, 2.00, 4.00, 2.01, 2.00, 2.50, 4.00, 3.23, 3.41, 3.00, 2.03, 2.23, 2.00, 5.16, 9.00, 2.50],
  "gender": ["Male", "Male", "Male", "Female", "Female", "Male", "Male", "Female", "Female", "Female", "Male", "Female", "Male", "Male", "Male", "Female", "Male", "Male", "Male", "Female"],
  "day": ["Thur", "Thur", "Thur", "Thur", "Thur", "Fri", "Fri", "Fri", "Fri", "Fri", "Sat", "Sat", "Sat", "Sat", "Sat", "Sun", "Sun", "Sun", "Sun", "Sun"],
  "time": ["Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner"],
  "size": [2, 2, 2, 4, 2, 2, 3, 2, 2, 2, 4, 3, 3, 4, 2, 2, 3, 4, 4, 2]
}
tip_dataframe = pd.DataFrame(tip_data_dict)
print(tip_dataframe.sort_values(by = ['gender', 'day']))
```

```python
sns.barplot(x="day", y="tip", hue="gender", data=tip_dataframe, errorbar=None, palette="deep")
plt.title("Average tip by day and gender")
plt.show()

#print(tip_dataframe.groupby(['gender', 'day'])['tip'].agg('mean'))
```

```python
sns.countplot(x="gender", data=tip_dataframe,hue = "day")
plt.title("count by gender")
plt.show()

#print(tip_dataframe.groupby('gender').count())
```

```python
sns.scatterplot(data=tip_dataframe, x="total_bill", y="tip",  hue = 'gender')
plt.title("total_bill vs tip")
plt.show()

#print(tip_dataframe.sort_values(by = ['total_bill', 'tip']))
```

```python
sns.boxplot(x='tip',  y = 'day', hue = 'gender', data = tip_dataframe)
plt.show()
```

```python
sns.histplot(data=tip_dataframe, x="total_bill",binwidth = 5)
plt.title("Bin-wise total bills")
plt.show()
```

```python
sns.histplot(data=tip_dataframe, x="total_bill", kde = True)
plt.title("kith KDE Enabled")
plt.show()
```

```python
sns.kdeplot(data=tip_dataframe, x="tip", fill = True)
plt.title("KDE plot")
plt.show()
```

```python
sns.violinplot(data = tip_dataframe, x = 'time', y='total_bill')
plt.title("total_bill Distribution by time (Lunch and Dinner)")
plt.show()
```
