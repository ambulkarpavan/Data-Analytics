```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
def lineplot(n=3):
    x = np.linspace(0, 10, 100)  # 100 values from 0  10
    for i in range(1, n + 1):
        y = x * i   # simple linear function, slope changes with i
        plt.plot(x, y, label=f"Line {i}")  # plot line
sns.set_theme()   
lineplot()


plt.legend()
plt.show()
sns.reset_orig()  # Reset back to Matplotlib
lineplot()
plt.title("Matplotlib defaults after sns.reset_orig()")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pal = "deep"
#pal = "pastel"
#pal = "muted"
#pal = "bright"
#pal = "dark"
#pal = "colorblind"

x=["A", "B", "C", "D"]
y=[4, 7, 3, 8]
sns.set_palette(pal)

plt.figure()
sns.barplot(x= x, y=y, hue = x)
plt.title(f"Palette = {pal}")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

pal = "deep"
sns.set_palette(pal)

# pick 4 colors from the palette explicitly
colors = sns.color_palette("bright", 4)

plt.figure()
sns.barplot(x=["A", "B", "C", "D"], y=[4, 7, 3, 8], palette=colors)
plt.title(f"Palette = {pal}")
plt.show()
```

```python
import seaborn as sns
import pandas as pd

# Create the dataframe
tip_data_dict = {
  "total_bill": [16.58, 7.56, 10.34, 43.11, 13.00, 13.51, 18.71, 12.74, 13.00, 16.40, 20.53, 16.47, 26.59, 38.73, 24.27, 12.76, 30.06, 25.89, 48.33, 13.27],
  "tip": [4.00, 1.44, 2.00, 5.00, 2.00, 2.00, 4.00, 2.01, 2.00, 2.50, 4.00, 3.23, 3.41, 3.00, 2.03, 2.23, 2.00, 5.16, 9.00, 2.50],
  "gender": ["Male", "Male", "Male", "Female", "Female", "Male", "Male", "Female", "Female", "Female", "Male", "Female", "Male", "Male", "Male", "Female", "Male", "Male", "Male", "Female"],
  "day": ["Thur", "Thur", "Thur", "Thur", "Thur", "Fri", "Fri", "Fri", "Fri", "Fri", "Sat", "Sat", "Sat", "Sat", "Sat", "Sun", "Sun", "Sun", "Sun", "Sun"],
  "time": ["Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner"],
  "size": [2, 2, 2, 4, 2, 2, 3, 2, 2, 2, 4, 3, 3, 4, 2, 2, 3, 4, 4, 2]
}
tip_dataframe = pd.DataFrame(tip_data_dict)
print(tip_dataframe.sort_values(by = ['gender', 'day']))
sns.barplot(x="day", y="tip", hue="gender", data=tip_dataframe, errorbar=None)
plt.title("Average tip by day and gender")
plt.show()
```

```python
sns.countplot(x="gender", data=tip_dataframe)
plt.title("count by gender")
plt.show()

print(tip_dataframe.groupby('gender')['day'].count())
```

```python
sns.countplot(x="gender", hue = "time", data=tip_dataframe)
plt.title("count by gender and day")
plt.show()

print(tip_dataframe.groupby(['gender','time'])['day'].count())
```

```python
sns.scatterplot(data=tip_dataframe, x="total_bill", y="tip")
plt.title("total_bill vs tip")
plt.show()

print(tip_dataframe.sort_values(by = ['total_bill', 'tip']))
```
