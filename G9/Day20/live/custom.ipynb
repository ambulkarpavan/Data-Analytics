```python
import matplotlib.pyplot as plt
import numpy as np

years = np.arange(2015, 2025)
admissions = np.array([45, 52, 48, 60, 75, 68, 82, 77, 90, 72])

# Find the highest point and lowest point
year_max, val_max = years[admissions.argmax()], admissions[admissions.argmax()]
year_min, val_min = years[admissions.argmin()], admissions[admissions.argmin()]

# Create plot in form of line chart
plt.subplots(figsize=(10, 5))
plt.plot(years, admissions, marker='o', linestyle='-', linewidth=2, color='tab:blue', label='Admissions')

# Annotate highest point (green arrow and green text)
plt.annotate(
    f"Highest\n{year_max}: {val_max}",
    xy=(year_max, val_max),
    xytext=(year_max, val_max + 10),
    arrowprops=dict(facecolor='green', edgecolor='green', arrowstyle='->', linewidth=1.5),
    color='green',
    fontsize=10,
    ha='center'
)
plt.annotate(
    f"Lowest\n{year_min}: {val_min}",
    xy=(year_min, val_min),
    xytext=(year_min, val_min + 10),
    arrowprops=dict(facecolor='red', edgecolor='red', arrowstyle='->', linewidth=1.5),
    color='red',
    fontsize=10,
    ha='center'
)
for x, y in zip(years, admissions):
    # place text slightly to the right and up in data units
    plt.text(x - 0.2, y + 3.0, f"({x}, {y})", fontsize=8, color = 'maroon')


min_year, max_year = years.min(), years.max()
min_adm, max_adm   = admissions.min(), admissions.max()

plt.xlim(min_year - 1, max_year + 1)
plt.ylim(min_adm - 10, max_adm + 10)

plt.grid(True, linestyle='--', alpha=0.5)



# Labels, title, legend
plt.xlabel("Year")
plt.ylabel("Number of Admissions")
plt.title("Admissions per Year (20152024)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

countries = ["Argentina", "Sweden", "South Africa", "Spain", "China", "Uganda", "Japan", "India", 'Mauritius']
densities = [44, 67, 140, 250, 390, 640, 880, 1260, 1650]

# Create scatter plot with viridis colormap
#plt.figure(figsize=(10,6))
plt.subplots(figsize=(10,6))
sc = plt.scatter(countries, densities, c=densities, cmap="viridis", edgecolor="black", marker="o", s=200)

# Add colorbar
cbar = plt.colorbar(sc)
cbar.set_label("Population Density (per km)", fontsize=12)

#code to add density as text for marker
for index in range(len(densities)):
  plt.text(countries[index], densities[index] + 50, densities[index], fontsize=8, color = 'black', ha = 'center')
```

```python
import numpy as np
from matplotlib import colormaps

#Get the inbuilt object for 'viridis' colormap. Viridis is a continuous colormap, meaning it smoothly interpolates colors for any value between 0 and 1.
viridis = colormaps['viridis']

# 6 points between 0 and 1
values = np.linspace(0, 1, 10)  
#get list of colors from array
colors = [viridis(v) for v in values]

#create a figure
plt.figure(figsize=(8, 1))
for i, c in enumerate(colors):
    #fill a rectangular area with the color specified 
    #[i, i+1] is the x co-ordinates
    #0, 1 is the y co-ordinates
    plt.fill_between([i, i+1], 0, 1, color=c)

#from the text
    plt.text(i+0.5, 0.5, f"{values[i]:.1f}", ha="center", va="center", color="white", fontsize=12)

plt.axis("off")
#set the title of the plot
plt.title("Viridis values at [0.0 to 1.0]", fontsize=14)
plt.show()

```
