```python

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Delhi', 'Mumbai', 'Chennai']
}

dFrame1 = pd.DataFrame(data)
dFrame1.to_csv('students.csv', index = False)
print("Check the", os.getcwd(), "for the file students.csv" )
```

```python
data = {
    'CustomerID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, np.nan, 35, 40],  # Include a NaN value
    'City': ['Delhi', 'Mumbai', 'Chennai', None],  # None to simulate a missing value
    'Spending': [2500.5, 1800.0, 3200.0, 1500.0]
}

dFrame3 = pd.DataFrame(data)
print(dFrame3)

dFrame3.to_csv('customers.csv', index = False)
print("Check the", os.getcwd(), "for the file customers.csv, especially take care for the missing value" )
```

```python
dFrame4 = pd.read_csv('customers.csv',  on_bad_lines = 'skip')
print(dFrame4)
```

```python
print("\nThe data type for every column is\n", dFrame4.dtypes, sep = "")
```

```python
dFrame4 = dFrame4.astype(
    {
        'CustomerID': 'Int64',
        'Name': 'string',
        'Age': 'Int64',
        'City': 'string',
        'Spending': 'Float64'
    }
)
```

```python
print("\nThe dataframe read from the customers after changing the datatype is\n",dFrame4, sep = "")
print("\nThe data type for every column is\n", dFrame4.dtypes, sep = "")
```
