```python
dFrame1 = pd.DataFrame(
  {
    'age' : [5, 6, np.nan],
    'born' : [pd.NaT, pd.Timestamp('1939-05-27'), pd.Timestamp('1940-04-25')],
    'name' : ['Alfred', 'Batman', ''],
    'toy' : [None, 'Batmobile', 'Joker']
  }
)
print("The dataframe dFrame1 is\n", dFrame1, sep = "")
print("\nThe dataframe against the missing values is\n", dFrame1.isna(), sep = "")
print("\nThe dataframe against the missing values is\n", dFrame1.isnull(), sep = "")
print("\nThe dataframe against the missing values is\n", dFrame1.isnull().sum().sum(), sep = "")
```

```python
dFrame2 = pd.DataFrame(
  {
    "name": ['Alfred', 'Batman', 'Catwoman'],
    "toy": [np.nan, 'Batmobile', 'Bullwhip'],
    "born": [pd.NaT, pd.Timestamp("1940-04-25"), pd.NaT],
    "salary": [np.nan, np.nan, np.nan]
  }
)
print("The original dFrame2 is\n", dFrame2, sep = "")
dFrame4 = dFrame2.dropna(how='all', axis = 1)
print("\nThe dFrame4 is\n", dFrame4, sep = "")
```

```python
dFrame4 = pd.DataFrame(
  {
    'A': [1,3,5,np.nan],
    'B': [2,np.nan,4,8],
    'C': [1,5,np.nan,15],
  }
)
print("The dFrame4 is:\n", dFrame4, sep ="")
```
