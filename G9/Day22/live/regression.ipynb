```python
import numpy as np

x = np.array([16.58, 7.56, 10.34, 43.11, 13.00, 13.51, 18.71, 12.74, 13.00, 16.40, 20.53, 16.47, 26.59, 38.73, 24.27, 12.76, 30.06, 25.89, 48.33, 13.27])
y = np.array([4.00, 1.44, 2.00, 5.00, 2.00, 2.00, 4.00, 2.01, 2.00, 2.50, 4.00, 3.23, 3.41, 3.00, 2.03, 2.23, 2.00, 5.16, 9.00, 2.50])

# Linear fit (degree=1)
coef_lin = np.polyfit(x, y, deg=1)   # [slope, intercept]
# Quadratic fit (degree=2)
coef_quad = np.polyfit(x, y, deg=2)  # [a2, a1, a0]

print("Linear coeffs (slope, intercept):", coef_lin)
print("Quadratic coeffs (a2, a1, a0):", coef_quad)
```

```python

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create the dataframe
tip_data_dict = {
  "total_bill": [16.58, 7.56, 10.34, 43.11, 13.00, 13.51, 18.71, 12.74, 13.00, 16.40, 20.53, 16.47, 26.59, 38.73, 24.27, 12.76, 30.06, 25.89, 48.33, 13.27],
  "tip": [4.00, 1.44, 2.00, 5.00, 2.00, 2.00, 4.00, 2.01, 2.00, 2.50, 4.00, 3.23, 3.41, 3.00, 2.03, 2.23, 2.00, 5.16, 9.00, 2.50],
  "gender": ["Male", "Male", "Male", "Female", "Female", "Male", "Male", "Female", "Female", "Female", "Male", "Female", "Male", "Male", "Male", "Female", "Male", "Male", "Male", "Female"],
  "day": ["Thur", "Thur", "Thur", "Thur", "Thur", "Fri", "Fri", "Fri", "Fri", "Fri", "Sat", "Sat", "Sat", "Sat", "Sat", "Sun", "Sun", "Sun", "Sun", "Sun"],
  "time": ["Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Lunch", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner", "Dinner"],
  "size": [2, 2, 2, 4, 2, 2, 3, 2, 2, 2, 4, 3, 3, 4, 2, 2, 3, 4, 4, 2]
}
tip_dataframe = pd.DataFrame(tip_data_dict)
print(tip_dataframe.sort_values(by = ['gender', 'day']))


sns.regplot(x="total_bill", y="tip", data=tip_dataframe)
plt.title("the Reg-Plot")
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=tip_dataframe, order = 2);
plt.title("the Reg-Plot")
plt.show()
```

```python
g = sns.FacetGrid(tip_dataframe, col="time",hue="gender" )
print(g)
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
```

```python
sns.lmplot(data = tip_dataframe, x="total_bill", y="tip", hue = 'gender')
plt.title("The lmplot")
plt.show()
```
